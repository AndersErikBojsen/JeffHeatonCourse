```python
def encode_numeric_zscore(df,name,mean=None,sd=None):
    if mean is None:
        mean = df[name].mean()

    if sd is None:
        sd = df[name].std()

    df[name] = (df[name]-mean)/sd
```

```python
encode_numeric_zscore(mpg,'mpg')
encode_numeric_zscore(mpg,'displacement')
encode_numeric_zscore(mpg,'weight')
encode_numeric_zscore(mpg,'mpg')
encode_numeric_zscore(mpg,'acceleration')
#transforma a coluna de object para int64 se não nao funciona
mpg['horsepower'] = mpg['horsepower'].astype('int')
encode_numeric_zscore(mpg,'horsepower')
print(mpg.mpg.mean())
print(mpg.mpg.std())
mpg.to_csv("/home/christian/data_analyse/JeffHeaton/JeffHeatonCourse/jeffheatoncourse/exercise_2.csv",index=False)
```

```python
#Exercise 4
mpg = pd.read_csv('/home/christian/data_analyse/JeffHeaton/JeffHeatonCourse/jeffheatoncourse/auto-mpg.csv')

x = mpg.iloc[:,1:]
y = mpg.iloc[:,0:1]
# Split into train/test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.23, random_state=42) 

#Train set
aux = y_train
aux['set'] = 'T'
T = pd.concat([aux,x_train],axis=1)

#Test set
aux = y_test
aux['set'] = 'V'
V = pd.concat([aux,x_test],axis=1)

#Final result
result = pd.concat([T,V],axis=0)
result.to_csv("/home/christian/data_analyse/JeffHeaton/JeffHeatonCourse/jeffheatoncourse/exercise_4.csv",index=False)

print("Training DF:", len(T.index))
print("Validation DF", len(V.index)) 


```
